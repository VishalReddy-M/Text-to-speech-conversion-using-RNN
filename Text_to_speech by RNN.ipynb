```python
import numpy as np
import tensorflow as tf
import subprocess
subprocess.check_call(['pip', 'install', 'gtts'])
from gtts import gTTS
import os
```

```python
text = " Learning RNN in a real world practical implementation" #this is a training sentence and RNN learns character patterns from this
#character encoding
# computers cannot understand the letters so we have to convert the characters into numbers
char_to_int = {c:i for i,c in enumerate(set(text))}
int_to_char = {i:c for c,i in char_to_int.items()} 
```

```python
sequence = [char_to_int[c] for c in text]
#creating input and output sequence
x = np.array(sequence[:-1]).reshape(1,-1,1)
y = np.array(sequence[:-1]).reshape(1,-1)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(None,1)),
    tf.keras.layers.SimpleRNN(32,return_sequences=True),
    tf.keras.layers.Dense(len(char_to_int),activation="softmax")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(x,y,epochs=200,verbose=1)
```

```python
speech_output = gTTS(text=text,lang="en")
```

```python
speech_output.save("speech.mp3")
os.system("start speech.mp3")
```
